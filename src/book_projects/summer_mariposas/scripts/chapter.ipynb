```python
import os
input_path = "../data/summer_of_the_mariposas.txt"
# Read the file into lines
with open(input_path, 'r', encoding='utf-8') as file:
    lines = file.read().splitlines()

chapter_names = [
    'EL CAZO', 'LA CALAVERA', 'EL PÁJARO', 'LA ESTRELLA', 'EL VENADO', 'EL MUNDO',
    'LA MANO', 'EL ÁRBOL', 'LA SIRENA', 'LA ARAÑA', 'LA GARZA', 'EL ALACRÁN',
    'LA MUERTE', 'LAS JARAS', 'EL DIABLITO', 'LA DAMA', 'EL NOPAL', 'LA CHALUPA',
    'EL CORAZÓN', 'EL MÚSICO', 'LA ROSA', 'LA CORONA', 'LA LUNA'
]
chapters = {}  # Dictionary to hold chapter texts
curr_chap = None  # Current chapter name
curr_text = []  # Current chapter text
for line in lines:
    # Determine if the line contains any chapter title
    found_chapter = None
    for chap_name in chapter_names:
        if chap_name in line:
            found_chapter = chap_name
            break
    if found_chapter:
        # If we were already accumulating a chapter, save it before starting a new one
        if curr_chap is not None:
            chapters[curr_chap] = '\n'.join(curr_text)
            curr_text = []
        curr_chap = found_chapter
        print(f"Starting new chapter: {curr_chap}")  # Debugging: Identify chapter starts
    elif curr_chap is not None:
        # Continue accumulating text for the current chapter
        curr_text.append(line)
# Save the last chapter if there was one
if curr_chap is not None and curr_text:
    chapters[curr_chap] = '\n'.join(curr_text)
print(f"Processed {len(chapters)} chapters.")
if len(chapters) == 0:
    print("No chapters were processed. Please check the chapter names and text formatting.")
else:
    for chap in chapters:
        print(f"Chapter '{chap}' has {len(chapters[chap].splitlines())} lines.")
output_dir = "../data/chapters"  # Directory to save the chapter files
os.makedirs(output_dir, exist_ok=True)  # Create the directory if it doesn't exist

chapter_num = 0

for chap_name, chap_text in chapters.items():
    # Create a safe filename by replacing spaces with underscores
    filename = f"{chapter_num} - {chap_name.replace(' ', '_')}.txt"
    chapter_num += 1

    filepath = os.path.join(output_dir, filename)
    with open(filepath, 'w', encoding="utf-8") as file:
        file.write(chap_text)
    print(f"Wrote chapter '{chap_name}' to file '{filepath}'")
```
